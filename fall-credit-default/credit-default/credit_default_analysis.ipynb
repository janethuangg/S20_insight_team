# dealing with categorical variables

```python
credit.columns
```

```python
y = credit.defaulted
features = credit.drop(columns=['defaulted'])
```

```python
categoricals = ['gender','education','marital_status','repay_status_sep', 'repay_status_aug', 'repay_status_jul',
       'repay_status_jun', 'repay_status_may', 'repay_status_apr']

for variable in categoricals:
    features[variable]=pd.Categorical(features[variable])
    dummies = pd.get_dummies(features[variable],prefix=variable, drop_first=True)
    features = pd.concat([features,dummies],axis=1)
    features.drop([variable],axis=1,inplace=True)
```

```python
features.head()
```

# linear regression

```python
X_mult = features
y_mult = y

mult_linreg = LinearRegression()
mult_linreg.fit(X_mult, y_mult)
```

```python
y_mult_pred = mult_linreg.predict(X_mult)
outcome_pred_class = np.where(y_mult_pred>=.5,1,0)

print("R^2: ", metrics.r2_score(y_mult, y_mult_pred))
print("Accuracy without splitting data:",metrics.accuracy_score(y_mult,outcome_pred_class))
```

```python
X_mult_train, X_mult_test, y_mult_train, y_mult_test = train_test_split(X_mult, y_mult, test_size=0.25, random_state=1)

#train on training set
mult_linreg2 = LinearRegression()
mult_linreg2.fit(X_mult_train, y_mult_train)

#generate predictions on training set and evaluate
y_mult_pred_train = mult_linreg2.predict(X_mult_train)
outcome_pred_train = np.where(y_mult_pred_train>=.5,1,0)
print("Accuracy on training data:",metrics.accuracy_score(y_mult_train,outcome_pred_train))

#generate predictions on test set and evaluate
y_mult_pred_test = mult_linreg2.predict(X_mult_test)
outcome_pred_test = np.where(y_mult_pred_test>=.5,1,0)
print("Accuracy on test data:",metrics.accuracy_score(y_mult_test,outcome_pred_test))
print("R^2: ", metrics.r2_score(y_mult_test, y_mult_pred_test))
```

# logistic regression

```python
X_mult_train, X_mult_test, y_mult_train, y_mult_test = train_test_split(X_mult, y_mult, test_size=0.25, random_state=1)

#train on training set
mult_logreg2 = LogisticRegression()
mult_logreg2.fit(X_mult_train, y_mult_train)

#generate predictions on training set and evaluate
y_mult_pred_train = mult_logreg2.predict(X_mult_train)
print("Accuracy on training data:",metrics.accuracy_score(y_mult_train,y_mult_pred_train))

#generate predictions on test set and evaluate
y_mult_pred_test = mult_logreg2.predict(X_mult_test)
print("Accuracy on test data:",metrics.accuracy_score(y_mult_test,y_mult_pred_test))
```

# decision tree

```python
t = DecisionTreeClassifier(max_depth=1, criterion='entropy')
t.fit(features,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2)

depths = [1,2,3,4,5,10,15,20,25,30]
accs = []
for i in depths:
    dtc = DecisionTreeClassifier(max_depth=i)
    dtc.fit(X_train,y_train)
    accs.append(dtc.score(X_test,y_test))
    
print('Scores:\n',pd.DataFrame({'Accuracy':accs},index=depths))
```

```python
dtc = DecisionTreeClassifier(max_depth=4,min_samples_leaf=10,min_samples_split=2,criterion='gini')
dtc.fit(X_train,y_train)

aucscore = metrics.roc_auc_score(dtc.predict(X_test), y_test)

fpr, tpr, thresholds = metrics.roc_curve(dtc.predict(X_test), y_test)

plt.plot(fpr, tpr, color='red',lw=2, label='ROC curve')
plt.plot([0, 1], [0, 1], color='blue', lw=2, linestyle='--')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve AUC = '+ str(aucscore.round(4)))
plt.show()
```

```python
test_cf = confusion_matrix(y_test,dtc.predict(X_test))
test_cf = test_cf.astype('float') / test_cf.sum(axis=1)[:, np.newaxis]
test_cf = pd.DataFrame((test_cf))
                           
fig, ax = plt.subplots(figsize=(6,6))
ax = sns.heatmap(test_cf,annot=True,cmap='Blues',cbar=False,fmt='.0%')
ax.set(xlabel='Predicted Class',ylabel='Actual Class',title='Normalized Confusion Matrix')
```

```python
dtc = DecisionTreeClassifier()
grid = {'max_depth':[2,5,10,20],
        'min_samples_leaf':[10,20,30],
        'min_samples_split':[2,3,4]}
gs = GridSearchCV(dtc,grid,cv=5,verbose=True,return_train_score=False)
gs.fit(features,y)

dtc_best = DecisionTreeClassifier(max_depth=5,min_samples_leaf=10,min_samples_split=3)
dtc_best.fit(X_train,y_train)


scores = pd.DataFrame(gs.cv_results_).filter(regex='param_+|mean_test_score'
                                            ).sort_values('mean_test_score',
                                                          ascending=False).reset_index().drop(['index'],axis=1)

scores.head(20)
```
